## Tipos de Dados


Tipos de Dados, usando como exemplo um veículo:
    
    1. Inteiro (ano de fabricação, quantidade de portas, largura, assentos, etc..)
    2. Real (potência do motor, preço, etc.. todo valor que contenha casa decimal)
    3. String (marca, modelo, tipo de carroceria, cor, etc... )
    4. Booleano ou tipo lógico (sim ou não, tem ou não tem, etc...)

No Python existes tipos de dados simples e estruturas.

Os dados simples são divididos em:
    1. Numéricos:
        a. Inteiro é int.
        b. Ponto flutuante ou real é float.

    2. Não numéricos:
        a. Lógico ou booleano é bool.
        b. String ou Texto é str.


As estruturas são divididas em :
  
    a. Lista é list
    b. Tupla é tuple
    c. Faixa é range
    d. Dicionário é dict

### Conversão de tipos

Em alguns momentos é necessário realizar a conversão de um tipo de dado para outro. Esse caso é chamado de *type casting* e é realizado pelo python conforme o exemplo abaixo:

```python
### conversão de um inteiro para uma string 
### variavelNova = str(variavel ou valor)

int_string = str(_funcionarioTotal)
type(int_string)
```

```python
int_string2 = str(5)
type(int_string2)
```

Para converter em outros tipos de dados, basta usar a mesma lógica:

**Converte para inteiro:**
    
    *variavelNova = int(variavel ou valor)*

**Converte para float:**

    *variavelNova = float(variavel ou valor)*

**Converte para bool:**

    *variavelNova = bool(variavel ou valor)*
    
   Sendo que qualquer valor diferente de 0 vale **True**, e apenas 0 vale **False**

### Estrutura de Dados - Lista

Listas são coleções de elementos, e são semelhantes à vetores ou arranjos (nomes usados em outras Linguagens como C e C++) com a vantagem de poder ter elementos de diversos tipos de dados.

Cada elemento da lista fica armazenda em uma posição diferente, chamado de índice. Sendo que o primeiro elemento fica armazenado no índie 0 e o último elemento no índice n-1 (sendo n a quantidade de elementos da lista).

São dois índices importante, o em azul (índice crescente) ou em vermelho (índice descrecente). Dessa forma é possível acessar o elemento 10 passando o índice <span style = "color: blue">  4 </span>  
ou pelo índice
<span style = "color: red"> -5 </span>.

### Principais métodos de um objeto lista

Método | Descrição | Exemplo
:----- | :-------- | :------
append | Adiciona um elemento no final da lista | `lista.append(5)`
clear  | Apaga todos os elementos de uma lista | `lista.clear()`
copy | Retorna uma cópia dos elementos da lista | `copia = lista.copy()`
count | Retorna a quantidade de ocorrências de um elemento na lista | `qt = lista.count(5)`
extend | Adiciona os elementos de outra lista passada por parâmetro | `lista.extend(outra_lista)`
index | Retorna o índice do elemento passado por parâmetro (primeira posição) | `pos5 = lista.index(5)`
insert | Adiciona um elemento em uma posição passada por parâmetro (adiciona no final caso a posição não exista)| `lista.insert(3, "João")`
pop | Remove o elemento na posição passada por parâmetro (provoca um erro caso a posição não exista)| `elemento = lista.pop(3)`
remove | Remove o elemento passado por parâmetro (provoca um erro caso o elemento não exista)| `lista.remove(5)`
reverse | Inverte a ordem dos elementos de uma lista | `lista.reverse()`
sort | Ordena os objetos de uma lista | `lista.sort()`
 | | `lista.sort(reverse=True)`
